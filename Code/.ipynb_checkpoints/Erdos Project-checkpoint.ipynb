```python
# import packages
import pandas as pd 
import numpy as np
import seaborn  as sns
import matplotlib.pyplot as plt
# !pip install pandas-profiling
```

```python
# read in data
hd       = pd.read_csv("../Data/heart_2020_cleaned.csv")
hd_train = pd.read_csv("../Data/data_train.csv")
hd_test  = pd.read_csv("../Data/data_test.csv")
```

```python
for column in hd_train.columns:
    plt.figure(figsize=(14,4))    
    plt.hist(hd_train.loc[hd_train.HeartDisease=="No",column], color='b', alpha=0.5, label="No HD")
    plt.hist(hd_train.loc[hd_train.HeartDisease=="Yes",column], color='r', alpha=0.5, label="HD")
    plt.legend(fontsize=14)
    plt.title(column, fontsize=16)
    plt.show()
```

```python
# Make dummy variables for Heart Disease and Stroke
hd_train.loc[:,'YesHD'] = pd.get_dummies(hd_train['HeartDisease'])['Yes'].copy()
hd_train.loc[:,'YesStroke'] = pd.get_dummies(hd_train['Stroke'])['Yes'].copy()
hd_train.loc[:,'YesSmoking'] = pd.get_dummies(hd_train['Smoking'])['Yes'].copy()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_reg_Stroke = LogisticRegression()
log_reg_Smoking = LogisticRegression()
log_reg_PH = LogisticRegression()
```

```python
log_reg_Stroke.fit(hd_train[["YesStroke"]], hd_train[["YesHD"]])
log_reg_Smoking.fit(hd_train[["YesSmoking"]], hd_train[["YesHD"]])
log_reg_PH.fit(hd_train[["PhysicalHealth"]], hd_train[["YesHD"]])
```

```python
log_reg_Stroke.predict(hd_train[["YesStroke"]])
log_reg_Smoking.predict(hd_train[["YesSmoking"]])
log_reg_PH.predict(hd_train[["PhysicalHealth"]])
```

```python
log_reg_Stroke.predict_proba(hd_train[["YesStroke"]])
log_reg_Smoking.predict_proba(hd_train[["YesSmoking"]])
log_reg_PH.predict_proba(hd_train[["PhysicalHealth"]])
```

```python
# Plot figure 
plt.figure(figsize = (10,8))

# With classifications we have a new method
# predict_proba which returns the probability
# that an observation is a certain class.
plt.plot(np.linspace(0,1,1000),
            log_reg_Stroke.predict_proba(np.linspace(0,1,1000).reshape(-1,1))[:,1],
            'r--',linewidth=2.5,label = "Model Fit")
plt.scatter(hd_train[["YesStroke"]], hd_train[["YesHD"]],label = 'Training Data',alpha=.7)
# plt.legend(fontsize = 14,loc = 4)
plt.xlabel("Stroke",fontsize = 16)
plt.ylabel("p(X)",fontsize=16) 

plt.show()
```

```python
## set the cutoff
cutoff = 0.92

## store the predicted probabilities
y_prob = log_reg_Stroke.predict_proba(hd_train[["YesStroke"]])[:,1]

## assign the value based on the cutoff
y_train_pred = 1*(y_prob >= cutoff)

## print the accuracy
## input the accuracy after "is",
print("The training accuracy for a cutoff of",cutoff,"is", np.sum(y_train_pred == hd_train.YesHD)/len(hd_train.YesHD))

```
